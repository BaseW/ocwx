```python
import numpy as np
def cos_sim(vec1, vec2):
    dot = np.dot(vec1, vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    try:
        sim = dot / (norm1 * norm2)
    except:
        sim = 0
    return sim

cos_sim([1,1], [0, 1])
```

Q2.4.2で作成した講義-単語行列の各要素(講義 i の単語 j のTFij )にその単語のIDF値を掛 けたTFIDF値を要素とする行列を作成して返す lec_word_tfidf_matrix 関数を完成させ てください。作成した行列は変数 tfidf_matrix で受け取ります。
ここで、講義 i 、単語 j のTFIDF値は以下のように定義されます。 
TFIDF=TF_ij*log(すべての講義数 /単語jのDF)=TF_ij*log(len(courses)/df[j])
log の計算には np.log() を使用してよいです。


```python
def lec_word_tfidf_matrix(courses, vocab, df):
    tf_matrix = lec_word_matrix(courses, vocab)
    row_length = len(tf_matrix)
    columns_length = len(tf_matrix[0])
    print(row_length, columns_length)
    tfidf_matrix = np.zeros((row_length,  columns_length), dtype=float)
    for i in range(row_length):
        for j in range(columns_length):
            idf = np.log(row_length / (df[j] + 1))
            tfidf_matrix[i][j] = idf * tf_matrix[i][j]
    return tfidf_matrix


tfidf_matrix = lec_word_tfidf_matrix(courses, vocab, df)
print(tfidf_matrix)
```
