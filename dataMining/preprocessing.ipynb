```python
def find_outliers(input_array):
    q1, q3 = np.percentile(input_array, [25, 75])
    iqr = q3 - q1
    outliers_list = []
    for item in input_array:
        if (item < q1 - 1.5 * iqr) or (item > q3 + 1.5 * iqr):
            outliers_list.append(item)
    return outliers_list

find_outliers(np.array([30,39,-100,29,95,70,67,200,29,1000,56,45,68]))
```

```python
import math
def correlation(x, y):
    mean_x = x.mean()
    mean_y = y.mean()
    sum_x = 0
    sum_y = 0
    sum_xy = 0
    for i in range(len(x)):
        sum_x += (x[i] - mean_x) ** 2
        sum_y += (y[i] - mean_y) ** 2
        sum_xy += (x[i] - mean_x) * (y[i] - mean_y)
    var_x = math.sqrt(sum_x)
    var_y = math.sqrt(sum_y)
    cov = sum_xy
    return var_x * var_y / cov

correlation(np.array([-1,1]),np.array([ -1, 1]))
```
