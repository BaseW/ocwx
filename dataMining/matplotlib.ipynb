```python
data =[0, 1, 4, 9, 16]
x =[0, 1, 2, 3, 4] 
y =[0, 1, 2, 3, 4]

plt.plot(x,y, linestyle='--', color='blue', marker='o', label="linear") 
plt.plot(data, linestyle=':', color='green', marker='*', label="quad") 
plt.legend();
```

```python
data =[0, 1, 4, 9, 16]
x =[0, 1, 2, 3, 4] 
y =[0, 1, 2, 3, 4]

plt.plot(x,y, linestyle='--', color='blue', marker='o', label="linear") 
plt.plot(data, linestyle=':', color='green', marker='*', label="quad") 
plt.legend();

plt.title("My First Graph") #  plt.xlabel("x") #x
plt.ylabel("y") #y
plt.grid(True); #
```

```python
# x
x = np.arange(-np.pi, np.pi, 0.1)

# cos, sin, y 
plt.plot(x,np.cos(x)) 
plt.plot(x,np.sin(x))
plt.title("cos ans sin Curves") #  
plt.xlabel("x") #x
plt.ylabel("y") #y
plt.grid(True); #
```

```python
x = np.arange(-np.pi, np.pi, 0.5) 
plt.plot(x,np.cos(x), marker='o') 
plt.plot(x,np.sin(x), marker='o') 
plt.title("cos ans sin Curves") 
plt.xlabel("x")
plt.ylabel("y") 
plt.grid(True);
```
